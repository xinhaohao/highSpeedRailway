```python
import numpy as np
from pandas import Series,DataFrame
import pandas as pd
import matplotlib.pyplot as plt
```

```python
d = pd.read_csv("../data/kalman_res.csv")
plt.figure(figsize=(10,6))
plt.scatter(d.index, d.iloc[:, 0], s=0.3)
plt.scatter(d.index, d.iloc[:, 3], s = 0.3)
my_yticks = np.array([-0.05,-0.02,-0.01,-0.005,0.005,0.01,0.02,0.05])
plt.yticks(my_yticks)
plt.grid()
plt.savefig("2.png")
plt.figure(figsize=(10,6))
ax1 = plt.subplot(2,1,1)
ax2 = plt.subplot(2,1,2)
ax1.scatter(d.index, d.iloc[:, 6], s=0.3)
ax2.scatter(d.index, d.iloc[:, 7], s = 0.3)
ax1.grid()
ax2.grid()
```

```python
file = "../data/diff.csv"
d = pd.read_csv(file)

fig = plt.figure(figsize=[10,8])
ax = fig.add_subplot(111)
ax2 = ax.twinx()
ax.plot(d.iloc[:, 0])
ax2.plot(d.iloc[:, 8], color='r')
ax2.plot(d.iloc[:, 7], color='g')
plt.savefig("3.png")
# ax2.scatter(np.arange(d.shape[0]),d.iloc[:, 6], color='r',s=0.1)
```
